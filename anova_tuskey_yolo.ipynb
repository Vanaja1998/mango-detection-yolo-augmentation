# Test results

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# -----------------------------
# 1. Create DataFrame
# -----------------------------

data = {
    "Model": ["YOLOv5","YOLOv5","YOLOv5",
        "YOLOv8","YOLOv8","YOLOv8",
        "YOLOv9","YOLOv9","YOLOv9",
        "YOLOv10","YOLOv10","YOLOv10",
        "YOLO11","YOLO11","YOLO11",
        "YOLO12","YOLO12","YOLO12"
    ],
    "D1": [ 0.897,
0.900,
0.889,0.895,0.899,0.88, 0.888,0.89,0.9, 0.89,0.9,0.89,
           0.886,0.91,0.899, 0.895,0.92,0.899],
    "D2": [  0.958,
0.960	,
0.955,0.966,0.96,0.967, 0.963,0.967,0.959, 0.956,0.942,0.955,
           0.955,0.945,0.952, 0.95,0.94,0.945],
    "D3": [0.950,
0.945,
0.952,0.961,0.956,0.966, 0.957,0.961,0.953, 0.953,0.951,0.952,
           0.937,0.936,0.94, 0.93,0.932,0.931]
}

df = pd.DataFrame(data)

# -----------------------------
# 2. Convert to Long Format
# -----------------------------

df_long = df.melt(id_vars=["Model"],
                  value_vars=["D1","D2","D3"],
                  var_name="Augmentation",
                  value_name="mAP")

# -----------------------------
# 3. Two-Way ANOVA
# -----------------------------

model = ols('mAP ~ C(Model) + C(Augmentation) + C(Model):C(Augmentation)',
            data=df_long).fit()

anova_table = sm.stats.anova_lm(model, typ=2)
print("Two-Way ANOVA Results")
print(anova_table)

# -----------------------------
# 4. Tukey HSD for Model Effect
# -----------------------------

tukey_model = pairwise_tukeyhsd(
    endog=df_long["mAP"],
    groups=df_long["Model"],
    alpha=0.05
)

print("\nTukey HSD - Model Effect")
print(tukey_model)

# -----------------------------
# 5. Tukey HSD for Augmentation Effect
# -----------------------------

tukey_aug = pairwise_tukeyhsd(
    endog=df_long["mAP"],
    groups=df_long["Augmentation"],
    alpha=0.05
)

print("\nTukey HSD - Augmentation Effect")
print(tukey_aug)
```
